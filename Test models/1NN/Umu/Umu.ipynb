```python
import numpy as np
import pandas as pd
from scipy.fft import fftn, ifftn, fftshift
import re

import matplotlib.pyplot as plt

def get_data(U,mu):
    '''
    reads a FRG output text file
    and returns the number of k1 and k2 points used, as well as the data for all three action channels

    '''
  


    df = pd.read_csv("output_" + f"{U}" +"_"+ f"{mu}" + "_40_15.txt", delimiter="\t", header = None, skiprows = 1) 

    r = df.iloc[-1, 0]
    n = r.split()

    tc, sc, mag, cdw, _ = map(float, n)

    orders = [sc, mag, cdw]
    o_max = max(orders)

    dominant = orders.index(o_max)

    


    return (tc, sc, mag, cdw, dominant)
```

```python
Us= np.array([0.5,1.0,1.5,2.0])
ratios= np.array([0.5, 1.0, 1.5, 2.0])

X, Y = np.meshgrid(Us, ratios)
Tcs = np.zeros((4,4))
dom = np.zeros((4,4))

for i, u in enumerate(Us):
    for j, r in enumerate(ratios):
        mu = u * r
        Tcs[i][j] = get_data(u,mu)[0]
        dom[i][j] = get_data(u, mu)[4]
```

```python
#plot tc


fig, ax = plt.subplots(figsize=(16, 6))

#ax.set_aspect('equal')
ax.grid()
C = ax.contourf(X, Y, Tcs,  levels = 30, cmap = 'coolwarm', linewidths=1.5)
fig.colorbar(C, ax=ax, orientation='vertical')

ax.set_title(f"Fig.1.6 1NN Tc as a function of  U and $\mu$ for nk= 40 and nkf= 15")
ax.set_ylabel("U")
ax.set_xlabel(r"$\frac{\mu}{U}$")
```
